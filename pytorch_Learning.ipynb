# to get informtion from Tensors - usally where the problem will arise while using tensors:


*   tensor not the right datatype -- use tensor.dtype
*   tensor not the right shape -- use tensor.shape
*   tensor not on the right device i.e(cpu,gpu) -- use tensor.device




```python
# create a tensor
some_tensor=torch.rand([3,2])
some_tensor
```

```python
# finding out information about the tensor -- attributes
print(f"the datatype of the tensor: {some_tensor.dtype}")
print(f"shape of the tensor: {some_tensor.shape}")
print(f"Device of the tensor: {some_tensor.device}")
```

```python
# to change the dtype:

some_tensor=some_tensor.type(torch.float16)
print(some_tensor)
```

TENSOR AGGREGATION
min,max,mean,sum

```python
x=torch.arange(0,100,10)
x
```

```python
#find min
torch.min(x),x.min()
```

```python
#find max
torch.max(x), x.max()
```

```python
#find mean - mean takes dtype as float cant use the int that we get when we do arange
torch.mean(x.type(torch.float32)), x.type(torch.float32).mean()
```

```python
#sum
torch.sum(x) , x.sum()
```

```python
#to find the index at which the min and max value occur:
print(x.argmin())
print(x.argmax())
```

RESHAPING, stacking, squeezing and unsqueezing(efficentnet?)

reshaping: to reshape the tensor
view: return the view of an input tesnro of certain shape but keep the same memory as the og tensor
stacking: stack diff tensors together
squeeze: removes all '1' dimesions from a tensor
unsqueeze : add a '1' dimension to a target tensor
permute: return a view of the input with dimensions permuted(swapped) a certain way

```python
# reshaping the tensor:

x=torch.arange(1,10)
x,x.shape
```

```python
x_reshaped=x.reshape(2,5)
x_reshaped , x_reshaped.shape # can only reshape by adding dimensions so that all the elements are still there i.e the product of the dimensions should always be the same
```

```python
# change the view:
z=x.view(3,3)
z,z.shape
```

```python
z[:, 0]=3 #### : means all rows 0: the index position in each row i.e that clm changes the value to 3
x, x.shape # the view and the actual tensor share memory so any change to the view will result in change to the tensor
```

```python
# stacking
x_stacked = torch.stack([x,x,x],dim=1) #hstack -> dim =0 stack the entire tensor on top of one another. vstack -> dim =1 each matching element combined and put in one row
x_stacked
```
